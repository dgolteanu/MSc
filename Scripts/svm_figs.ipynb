```python
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.datasets import make_blobs, make_gaussian_quantiles, make_circles
from sklearn.inspection import DecisionBoundaryDisplay

X, y = make_circles(n_samples=100,random_state=6, noise=0.1, factor=0.4)
# X,y = make_gaussian_quantiles(n_samples=100,n_features=2, n_classes=2,random_state=6)
# Quadratic SVM soft margin
# fit the model
clf = svm.SVC(kernel="poly",degree=2, C=100)
clf.fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(X[:, 0].min() - 0.1, X[:, 0].max() + 0.1)
ax.set_ylim(X[:, 1].min() - 0.1, X[:, 1].max() + 0.1)
# plt.savefig('svm_quad_soft_margin.pdf', format='pdf',bbox_inches='tight')
plt.show()
```

```python
# Linear SVM soft margin
# we create 40 separable points
X, y = make_blobs(n_samples=100, centers=[[-0.5, -0.5], [2.0, 2.0]],cluster_std = [1.0, 0.5], random_state=6)
# train
clf = svm.SVC(kernel="linear", C=0.1)
clf.fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
ax.set_xticks([])
ax.set_yticks([])
# plt.savefig('svm_soft_margin.pdf', format='pdf',bbox_inches='tight')
plt.show()
```
